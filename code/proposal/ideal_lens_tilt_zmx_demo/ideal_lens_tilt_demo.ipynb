## Target pattern

```python
# With hints from creating exact number of cycles in a linear sinusoidal signal

target_size = 12  # unit = cm
DPI = 200
dots = target_size*DPI/2.54

# ronchi or sinusoidal ?
ronchi = False

pattype = 'ronchi' if ronchi else 'sinusoid'

N = dots if np.mod(dots, 2) else dots + 1
figS = N/DPI

# pattern specification
lpmm_x = 3
lpmm_y = 0

cycles_x = 25.4*lpmm_x
cycles_y = 25.4*lpmm_y

x = np.linspace(0, N-1, N) - N//2
x = x/(N-1)
y = x.copy()
X, Y = np.meshgrid(x, y)
a = 0.5
b = 0.5
threshold = 0.5

phase_off = np.pi/32 # to prevent unequal width stripes
pat = a + b*np.sin(2*np.pi*cycles_x*X + 2*np.pi*cycles_y*Y + phase_off)
if ronchi:
    pat[pat < threshold] = 0.0
    pat[pat >= threshold] = 1.0

# plotting
def plot_sinusoid(target, filename, save=False):
    global figS, DPI
    fig = plt.figure(figsize=(figS,figS), dpi=DPI)
    ax = fig.add_axes([0.0,0.0,1.0,1.0])
    ax.imshow(target, 'gray', aspect='equal', interpolation='none')
    ax.set_aspect('equal')
    ax.xaxis.set_ticks([]); ax.yaxis.set_ticks([])
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    if save:
        fig.savefig(filename, dpi=DPI, bbox_inches='tight', pad_inches=0.00)
    plt.show()
   
# figure plotting and saving
imgFilename = "propDemo_{}_lpmmx{}_lpmmy{}.png".format(pattype, lpmm_x, lpmm_y)
zmxImgDir = os.path.join(ln.zGetPath()[0], "IMAFiles") 
filename = os.path.join(zmxImgDir, imgFilename)
plot_sinusoid(pat, filename, False)

# information about the pattern itself
print("image dir: ", zmxImgDir)
print("filename: ", imgFilename)
print("max value:", np.max(pat))
print("min value:", np.min(pat))
print("cycles (X):", cycles_x)
print("Samples/cycle (X): ", N/cycles_x)
print("Digitized frequency (X): ", cycles_x/N)
print("cycles (Y):", cycles_y)
#print("Samples/cycle (Y): ", N/cycles_y)
print("Digitized frequency (Y): ",  cycles_y/N)
hori_slice = pat[N//2,:]
print(len(hori_slice), len(hori_slice)/2.0)
print(np.sum(hori_slice))
```

#### PSFs and MTFs

```python
# Helper functions to plot PSFs and MTF
def plotPsf(psfinfo, psfdata):
    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(111)
    halfLen = psfinfo.dataArea/2
    ax.imshow(psfdata, cmap=psfcm, 
              extent=(-halfLen, halfLen, -halfLen, halfLen))
    ax.set_xlabel('microns')
    ax.set_ylabel('microns')
    plt.show()
    
def plotMtf(mtfs):
    fig = plt.figure(figsize=(10,6))
    ax = fig.add_subplot(111)

    for field, mtf in enumerate(mtfs):
        ax.plot(mtf.SpatialFreq, mtf.Tangential, label='F-{}, T'.format(field + 1))
        ax.plot(mtf.SpatialFreq, mtf.Sagittal, label='F-{}, S'.format(field + 1))
        ax.set_xlabel('Spatial Frequency in cycles per mm', fontsize=14)
        ax.set_ylabel('Modulus of the OTF', fontsize=14)
        ax.grid('on')
    ax.legend(frameon=False)
    plt.show()
    
def plotImageSim(img_info, img_data):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111)
    img_data_np = np.array(img_data, dtype='uint8')
    left, right = -img_info.imgW/2, img_info.imgW/2
    bottom, top = -img_info.imgH/2, img_info.imgH/2
    ax.imshow(img_data_np, extent=[left, right, bottom, top])
    plt.show()
```

## Scheimpflug camera configuration

Set the object surface back to the initial distance

```python
ln.zSetSurfaceData(1, 3, u0)
```

We would like to tilt the plane of sharp focus by 80 degrees (say, somewhat arbitrarily). We can get the corresponding tilt of the image plane using geometric calculations. We will set the computed image plane tilt in Zemax, and ask it to optimize for RMS spot size (using Default merit function) to find the plane of sharp focus angle.

<font color='red'>Also, it seems that (at least now), we need to help the optimization algorithm, by giving it a good starting value for the angle, if not, then it is not able to find the right value</font>

```python
cam = imgr.scheimpflug(f=f, u=u0)
```

```python
cam.v
```

```python
cam.lensPosfAng_d = 80
```

```python
cam.lensImgAng_d
```

#### Set image plane tilt

```python
ln.zSetSurfaceParameter(surfaceNumber=4, parameter=3, value=cam.lensImgAng_d)
```

```python
ln.zPushLens(1)
```

#### Place variable solve on object plane tilt for optimization

```python
surfNum, scode = 1, 7   # surf#1 (CB-1)'s parameter 3 (Tilt-X), parameter 1 start at 5 
ln.zSetSolve(surfNum, scode, 1, '') # variable solve
```

```python
ln.zPushLens(1)
```

Now, let try to optimize and see what Zemax reports for the tilt angle (we expect 80 degrees)

```python
print("Merit function before optimization:", ln.zOptimize(-1))
print("Merit function after optimization:", ln.zOptimize())
posfAngleAfterOpt = ln.zGetSurfaceParameter(surfaceNumber=1, parameter=3)
print("POSF angle after optimization", posfAngleAfterOpt, "degrees")
```

```python
ln.zSetSurfaceParameter(surfaceNumber=1, parameter=3, value=60)
```

```python
print("Merit function before optimization:", ln.zOptimize(-1))
print("Merit function after optimization:", ln.zOptimize())
posfAngleAfterOpt = ln.zGetSurfaceParameter(surfaceNumber=1, parameter=3)
print("POSF angle after optimization", posfAngleAfterOpt, "degrees")
```

That is pretty close to what we expect. We shall remove the variable on this parameter now.

```python
ln.zSetSolve(surfNum, scode, 0, '');  # remove the solve
```

```python
ln.zPushLens(1)
```

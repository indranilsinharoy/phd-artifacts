# Three dimensional light distribution simulation

Start with the Lommel function U

```python
import scipy.special as sps

def un(u, v, n, s=20):
    """"Lommel function Un, see def. in Born & Wolf, page 487, Edition 7
    
    Parameters
    ----------
    u : flattened array coordinate
    v : flattened array coordinate
    n : order of the lommel function Un
    s : order of the summation. 
    """
    uval = np.zeros(np.shape(u))         # The dtype of uval is 'float64'  
    for i in range(len(uval)):
        for j in range(s+1):
                uval[i] = uval[i] + ((-1)**j)*((u[i]/v[i])**(n+2*j))*sps.jn(n + 2*j, v[i])
    return uval

def vn(u, v, n, s=20):
    """Lommel function Vn, see def. in Born & Wolf, page 487, Edition 7
    
    Parameters
    ----------
    u : flattened array coordinate
    v : flattened array coordinate
    n : order of the lommel function Vn
    s : order of the summation.
    """
    vval = zeros(shape(u))         # The dtype of uval is 'float64'
    for i in range(len(vval)):
        for j in range(s+1):
            vval[i] = vval[i] + ((-1)**j)*((v[i]/u[i])**(n+2*j))*sps.jn(n + 2*j, v[i])
    return vval
```

```python
s = 20
ut = np.array([-10, -2,-1,0.009,1,2,20,20,300])
vt = np.array([-300, -2,-1,0.009,1,2,-10,10,10])
utt = np.linspace(-600,600, 1200)
vtt = np.linspace(-600,600, 1200)

# Expectation:
# for the first values of ut and vt, abs(vt/ut) > 1. Therefore vn will not converge (will diverge)
# for the last 3 values (especially the last values) of ut and vt, abs(ut/vt) > 1. Therefore un will not converge.


# Time
%timeit un(utt, vtt, 0, s)
%timeit vn(utt, vtt, 0, s)

# Values
uval_n0 = un(ut, vt, 0, s)
uval_n1 = un(ut, vt, 1, s)
uval_n2 = un(ut, vt, 2, s)
print "Un_n0:\n", uval_n0
print "Un_n1:\n", uval_n1
print "Un_n2:\n", uval_n2 

vval_n0 = vn(ut, vt, 0, s)
vval_n1 = vn(ut, vt, 1, s)
vval_n2 = vn(ut, vt, 2, s)
print "Vn_n0:\n", vval_n0
print "Vn_n1:\n", vval_n1
print "Vn_n2:\n", vval_n2
```

Although the performance has improved from 1 sec to 0.23 sec, can we improve the performance further?


Cython annotation (`cython -a xxx.pyx` @ cmd shell) suggest that there are still two main bottlenecks:

1.  The calculation of jn from the signal processing module 
2.  The division by v[i] because Cython is trying to check for divide by zero error.



The divide-by-zero exception check can be very expensive within large loops. If we can restrict that u, and v will not have zero values, then we can use the cython decorator `@cython.cdivision(True)` to avoid the exception checking. See [here](http://stackoverflow.com/questions/4709285/cython-float-division-pyexc-zerodivisionerror-checking) 

```python
%%cython
import numpy as np
import scipy.special as sps
import cython

@cython.wraparound(False)
@cython.boundscheck(False)
@cython.cdivision(True)
def un_cy02(double[::1] u, double[::1] v, int n, int s=20):
    """"Lommel function Un, see def. in Born & Wolf, page 487, Edition 7
    
    Parameters:
    ----------
    u : flattened array coordinate
    v: flattened array coordinate
    n: order of the lommel function Un
    s: order of the summation. 
    """
    cdef:
        int i, j, L
        double[::1] uval
        double jn, ui, vi
    
    L = len(u)        
    uval = np.zeros((L,), dtype=np.float64)         # The dtype of uval is 'float64'  

    # Calculate intensity at every point
    
    for i in range(L):
        for j in range(s+1):
                ui = u[i]
                vi = v[i]
                jn = sps.jn(n + 2*j, vi)
                uval[i] += ((-1)**j)*((ui/vi)**(n+2*j))*jn
    return np.asarray(uval)
```

```python
import numpy as np
import scipy.special as sps


lam = 550e-6                        # wavelength (lambda) in mm
a = 0.5                               # radius of the aperture in mm
f = 25                              # focal length in mm

# Global parameters
s = 16 
k = 2*pi/lam
k1 = k*(a/f)**2 # then u = k1*z
k2 = k*(a/f)    # then v = k2*sqrt(x^2 + y^2)


def un(u, v, n, s=20):
    """"Lommel function Un, see def. in Born & Wolf, page 487, Edition 7
    
    Parameters
    ----------
    u : flattened array coordinate
    v : flattened array coordinate
    n : order of the lommel function Un
    s : order of the summation. 
    """
    uval = np.zeros(np.shape(u))         # The dtype of uval is 'float64'  
    for i in range(len(uval)):
        for j in range(s+1):
                uval[i] = uval[i] + ((-1)**j)*((u[i]/v[i])**(n+2*j))*sps.jn(n + 2*j, v[i])
    return uval

def vn(u, v, n, s=20):
    """Lommel function Vn, see def. in Born & Wolf, page 487, Edition 7
    
    Parameters
    ----------
    u : flattened array coordinate
    v : flattened array coordinate
    n : order of the lommel function Vn
    s : order of the summation.
    """
    vval = zeros(shape(u))         # The dtype of uval is 'float64'
    for i in range(len(vval)):
        for j in range(s+1):
            vval[i] = vval[i] + ((-1)**j)*((v[i]/u[i])**(n+2*j))*sps.jn(n + 2*j, v[i])
    return vval


def func1(z, x, y, s):
    "Normalized intensity distribution for region |u/v| > 1"
    global k1, k2
    u = k1*z
    v = k2*sqrt(x**2 + y**2)
    vn0 = vn(u,v,0,s)
    vn1 = vn(u,v,1,s)
    return (((2/u)**2)*(1 + vn0**2 + vn1**2 -
                         2*vn0*np.cos(0.5*(u + (v**2 /u))) -
                         2*vn1*np.sin(0.5*(u + (v**2 /u)))))

def func2(z, x, y, s):
    "Normalized intensity distribution for region |u/v| < 1 "
    global k1, k2
    u = k1*z
    v = k2*sqrt(x**2 + y**2)
    un1 = un(u, v, 1, s)
    un2 = un(u, v, 2, s)
    return  ((2/u)**2)*(un1**2 + un2**2)


def func3(z, x, y):
    "Normalized intensity distribution for region |u/v| ~ 1"
    global k1,k2
    u = k1*z
    return (1 - 2*sps.jn(0,u)*np.cos(u) + (sps.jn(0,u))**2)/(u**2)


def intensityDistribution(z, x, y):
    global f, a, s
    #initialization
    intensity = np.zeros(np.shape(x))
    m1 = np.zeros(np.shape(x))        # masks that defines region 1
    m2 = np.zeros(np.shape(x))        # masks that defines region 2
    m3 = np.zeros(np.shape(x))        # masks that defines region 3

    # The bracketing parameters
    low_bracket = 0.999 #0.999
    up_bracket = 1 + (1-low_bracket)

    # Create the masks
    # Region 1: u > v or z > (f/a)*sqrt(x^2 + y^2)
    b = abs(z) > up_bracket*(f/a)*abs(sqrt(x**2 + y**2))
    m1[b==True]=1; #print 'm1\n',m1
    #print "done m1"
    del b;
    # Region 2: u < v or z < (f/a)*sqrt(x^2 + y^2)
    b = abs(z) < low_bracket*(f/a)*abs(sqrt(x**2 + y**2))
    m2[b==True]=1; #print 'm2\n',m2
    #print "done m2"
    del b;
    # Region 3: u approx equal v or z approx. equal to (f/a)*sqrt(x^2 + y^2)
    # for this we need to create two masks and then take an AND operation
    b1 = low_bracket*(f/a)*abs(sqrt(x**2 + y**2)) <= abs(z);
    b2 = abs(z) <= up_bracket*(f/a)*abs(sqrt(x**2 + y**2))
    b = b1 * b2  # I guess that this is the only way of doing a AND b (to check)
    m3[b==True]=1; #print 'm3\n',m3
    #print "done m3"
    del b,b1,b2

    #print shape(intensity)
    #print "sum(m1),sum(m2),sum(m3)",sum(m1),sum(m2),sum(m3); #return intensity
    #print "sum of intensity start", sum(intensity)
    #print "Calculating intensity in region 1"
    intensity[m1==1] = func1(z[m1==1],x[m1==1],y[m1==1], s)
    #print "sum of intensity after func1", sum(intensity)
    #print "Calculating intensity in region 2"
    intensity[m2==1] = func2(z[m2==1],x[m2==1],y[m2==1], s)
    #print "sum of intensity after func2", sum(intensity)
    #print "Calculating intensity in region 3"
    intensity[m3==1] = func3(z[m3==1],x[m3==1],y[m3==1])
    #print "sum of intensity after func3", sum(intensity)
    #print "shape_intensity", shape(intensity)
    return intensity
```

```python
%%cython

import numpy as np
from scipy.special import jn
cimport numpy as np
import cython
from libc.math cimport sin, cos, sqrt, abs

DTYPE = np.double   
ctypedef np.double_t DTYPE_t

cdef float a, f
cdef double lam, k, k1, k2, PI
cdef int s

#lam = 550e-6                          # wavelength (lambda) in mm
one_by_lambda = 1818.181818181818
a = 0.5                               # radius of the aperture in mm
f = 25.0                              # focal length in mm

# Global parameters
s = 16 
PI = 3.141592653589793
k = 2.0*PI*one_by_lambda
k1 = k*(a/f)**2.0  # then u = k1*z
k2 = k*(a/f)       # then v = k2*sqrt(x^2 + y^2)


@cython.wraparound(False)
@cython.boundscheck(False)
@cython.cdivision(True)
def un(np.ndarray[DTYPE_t, ndim=1] u, np.ndarray[DTYPE_t, ndim=1] v, int n, int s=20):
    """"Lommel function Un, see def. in Born & Wolf, page 487, Edition 7
    
    Parameters:
    ----------
    u : flattened array coordinate
    v: flattened array coordinate
    n: order of the lommel function Un
    s: order of the summation. 
    """
    cdef:
        int i, j, L
        double jnvi, ui, vi
    L = len(u)         
    cdef np.ndarray[DTYPE_t, ndim=1] uval = np.zeros((L,), dtype=DTYPE)

    # Calculate intensity at every point
    for i in range(L):
        for j in range(s+1):
                ui = u[i]
                vi = v[i]
                jnvi = jn(n + 2*j, vi)
                uval[i] = uval[i] + ((-1)**j)*((ui/vi)**(n+2*j))*jnvi
    return np.asarray(uval)

@cython.wraparound(False)
@cython.boundscheck(False)
@cython.cdivision(True)
def vn(np.ndarray[DTYPE_t, ndim=1] u, np.ndarray[DTYPE_t, ndim=1] v, int n, int s=20):
    """Lommel function Vn, see def. in Born & Wolf, page 487, Edition 7
    
    Parameters
    ----------
    u : flattened array coordinate
    v : flattened array coordinate
    n : order of the lommel function Vn
    s : order of the summation.
    """
    cdef:
        int i, j, L
        double jnvi, ui, vi
    L = len(u)         
    cdef np.ndarray[DTYPE_t, ndim=1] vval = np.zeros((L,), dtype=DTYPE)
    
    for i in range(L):
        for j in range(s+1):
            ui = u[i]
            vi = v[i]
            jnvi = jn(n + 2*j, vi)
            vval[i] = vval[i] + ((-1)**j)*((vi/ui)**(n+2*j))*jnvi
    return vval


def func1(np.ndarray[DTYPE_t, ndim=1] z, np.ndarray[DTYPE_t, ndim=1] x, np.ndarray[DTYPE_t, ndim=1] y, int s=20):
    "Normalized intensity distribution for region |u/v| > 1"
    # note this function will be called only once for each segment
    global k1, k2
    
    cdef int zL = z.shape[0]
    cdef np.ndarray[DTYPE_t, ndim=1] result = np.empty((zL,), dtype=DTYPE)   
    cdef np.ndarray[DTYPE_t, ndim=1] u = np.empty((zL,), dtype=DTYPE)
    cdef np.ndarray[DTYPE_t, ndim=1] v = np.empty((zL,), dtype=DTYPE)
    cdef np.ndarray[DTYPE_t, ndim=1] vn0, vn1

    
    cdef int i = 0
    for i in range(zL):
        u[i] = k1*z[i]
        v[i] = k2*sqrt(x[i]**2 + y[i]**2)
    
    vn0 = vn(u,v,0,s)
    vn1 = vn(u,v,1,s)
    
    for i in range(zL):
        result[i] = (((2.0/u[i])**2.0)*(1.0 + vn0[i]**2.0 + vn1[i]**2.0 -
                         2.0*vn0[i]*cos(0.5*(u[i] + (v[i]**2.0 /u[i]))) -
                         2.0*vn1[i]*sin(0.5*(u[i] + (v[i]**2.0 /u[i])))))
    return result

def func2(np.ndarray[DTYPE_t, ndim=1] z, np.ndarray[DTYPE_t, ndim=1] x, np.ndarray[DTYPE_t, ndim=1] y, int s=20):
    "Normalized intensity distribution for region |u/v| < 1 "
    # note this function will be called only once for each segment
    global k1, k2
    cdef np.ndarray[DTYPE_t, ndim=1] u, v
    cdef np.ndarray[DTYPE_t, ndim=1] un1, un2
    cdef np.ndarray[DTYPE_t, ndim=1] result
    
    u = k1*z
    v = k2*np.sqrt(x**2.0 + y**2.0)
    un1 = un(u, v, 1, s)
    un2 = un(u, v, 2, s)
    result = ((2.0/u)**2)*(un1**2.0 + un2**2.0)
    return  result


def func3(np.ndarray[DTYPE_t, ndim=1] z):
    "Normalized intensity distribution for region |u/v| ~ 1"
    # note this function will be called only once for each segment
    global k1
    #cdef np.ndarray[DTYPE_t, ndim=1] u
    cdef int zL = z.shape[0]
    cdef double u
    cdef np.ndarray[DTYPE_t, ndim=1] result = np.empty((zL,), dtype=DTYPE)
    cdef int i = 0
    for i in range(zL):
        u = k1*z[i]
        result[i] = (1.0 - 2.0*jn(0,u)*cos(u) + jn(0,u)**2.0)/(u**2.0)
    #u = k1*z
    #result = (1.0 - 2.0*jn(0,u)*np.cos(u) + jn(0,u)**2.0)/(u**2.0)
    return result


def intensityDistribution(np.ndarray[DTYPE_t, ndim=3] z, 
                          np.ndarray[DTYPE_t, ndim=3] x, 
                          np.ndarray[DTYPE_t, ndim=3] y):
    global f, a, s
    #initialization
    intensity = np.zeros(np.shape(x), dtype=DTYPE)
    m1 = np.zeros(np.shape(x), dtype=np.uint16)        # masks that defines region 1
    m2 = np.zeros(np.shape(x), dtype=np.uint16)        # masks that defines region 2
    m3 = np.zeros(np.shape(x), dtype=np.uint16)        # masks that defines region 3

    # The bracketing parameters
    low_bracket = 0.999 #0.999
    up_bracket = 1.0 + (1.0-low_bracket)
        
    # Create the masks  
    # Region 1: u > v or z > (f/a)*sqrt(x^2 + y^2)
    b = np.abs(z) > up_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2))
    m1[b==True] = 1  
    
    # Region 2: u < v or z < (f/a)*sqrt(x^2 + y^2)
    b = np.abs(z) < low_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2))
    m2[b==True] = 1 
 
    # Region 3: u approx equal v or z approx. equal to (f/a)*sqrt(x^2 + y^2)
    # for this we need to create two masks and then take an AND operation
    b1 = low_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2)) <= np.abs(z);
    b2 = np.abs(z) <= up_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2))
    b = b1 * b2   
    m3[b==True] = 1
 

    #print "Calculating intensity in region 1"
    intensity[m1==1] = func1(z[m1==1],x[m1==1],y[m1==1], s)

    #print "Calculating intensity in region 2"
    intensity[m2==1] = func2(z[m2==1],x[m2==1],y[m2==1], s)

    #print "Calculating intensity in region 3"
    intensity[m3==1] = func3(z[m3==1])

    return intensity
```

```python
lam = 550e-6
a = 0.5
f =25
u_div = [-5.0, 0.0] 
v_min = 1.0; v_max = 4.0
numdiv = len(u_div)

for i in range(len(u_div)-1):
    #print "\nSEGMENT NUMBER %d:" %(numdiv+i+1)
    #print "----------------"
    u_min = u_div[i]; u_max = u_div[i+1];
    
    #Calculate corresponding z_min,z_max, x_min, x_max, y_min and y_max (in mm)
    z_min = (u_min*lam*f**2)/(2*pi*a**2)
    z_max = (u_max*lam*f**2)/(2*pi*a**2)
    x_min = (v_min*lam*f)/(2*pi*a)
    x_max = (v_max*lam*f)/(2*pi*a)
    y_min = (v_min*lam*f)/(2*pi*a)
    y_max = (v_max*lam*f)/(2*pi*a)

    nXY = 4j
    nZ  = 6j  
    # Create the grid
    x, y, z = np.mgrid[x_min:x_max:nXY, y_min:y_max:nXY, z_min:z_max:nZ]
    # To ensure that we don't have any zeros
    # Need to do this because this is an assumption within the fast cythonized Un and Vn calculations.
    x[x==0.0]=10e-16
    y[y==0.0]=10e-16
    z[z==0.0]=10e-16



#initialization
intensity = np.zeros(np.shape(x))
m1 = np.zeros(np.shape(x))        # masks that defines region 1
m2 = np.zeros(np.shape(x))        # masks that defines region 2
m3 = np.zeros(np.shape(x))        # masks that defines region 3

# The bracketing parameters
low_bracket = 0.999 #0.999
up_bracket = 1.0 + (1.0-low_bracket)

# Flag
no_for_loop = True

# Create the masks 
if no_for_loop:
    print "Not using for loop"
    # Region 1: u > v or z > (f/a)*sqrt(x^2 + y^2)
    bm = np.abs(z) > up_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2))
    m1[bm==True] = 1
    
    # Region 2: u < v or z < (f/a)*sqrt(x^2 + y^2)
    b = np.abs(z) < low_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2))
    m2[b==True] = 1 
    
    # Region 3: u approx equal v or z approx. equal to (f/a)*sqrt(x^2 + y^2)
    # for this we need to create two masks and then take an AND operation
    b1 = low_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2)) <= np.abs(z)
    b2 = np.abs(z) <= up_bracket*(f/a)*np.abs(np.sqrt(x**2 + y**2))
    b = b1 * b2   
    m3[b==True] = 1
else:
    print "Using For loop"
    # Using for looop
    bt = np.zeros(np.shape(x))
    ix_N = bt.shape[0]
    iy_N = bt.shape[1]
    iz_N = bt.shape[2]
    ubf_by_a =  up_bracket*(f/a)
    lbf_by_a =  low_bracket*(f/a)
    
    # Create abs(sqrt(x**2 + y**2)) matrix
    ab_sqrt_x2plusy2 = np.zeros(np.shape(x))
    for ix in range(ix_N):
        for iy in range(iy_N):
            for iz in range(iz_N):
                ab_sqrt_x2plusy2[ix, iy, iz] = abs(sqrt(x[ix, iy, iz]**2 + y[ix, iy, iz]**2))
    
    # Region 1: u > v or z > (f/a)*sqrt(x^2 + y^2)
    for ix in range(ix_N):
        for iy in range(iy_N):
            for iz in range(iz_N):
                if (abs(z[ix, iy, iz]) > ubf_by_a*ab_sqrt_x2plusy2[ix, iy, iz]):
                    bt[ix, iy, iz] = 1
    m1[bt==1] = 1
    
    # Region 2: u < v or z < (f/a)*sqrt(x^2 + y^2)
    # reset bt
    bt = np.zeros(np.shape(x))
    for ix in range(ix_N):
        for iy in range(iy_N):
            for iz in range(iz_N):
                if (abs(z[ix, iy, iz]) < lbf_by_a*ab_sqrt_x2plusy2[ix, iy, iz]):
                    bt[ix, iy, iz] = 1
    m2[bt==1] = 1 
                    
    # Region 3: u approx equal v or z approx. equal to (f/a)*sqrt(x^2 + y^2)
    # reset bt
    bt = np.zeros(np.shape(x))
    for ix in range(ix_N):
        for iy in range(iy_N):
            for iz in range(iz_N):
                zval = z[ix, iy, iz]
                sqrtxyval = ab_sqrt_x2plusy2[ix, iy, iz]
                if ((lbf_by_a*sqrtxyval <= abs(zval)) and 
                    (abs(zval) <= ubf_by_a*sqrtxyval)):
                    bt[ix, iy, iz] = 1
    m3[bt==1] = 1 


intensity[m1==1] = 1
intensity[m2==1] = 2
intensity[m3==1] = 3

intensity


#print x.shape
#print x

#print ix_N, iy_N, iz_N

#print "bm"
#print bm.astype(np.int)
#print "bt"
#print bt.astype(np.int)
#print "bt - bm"
#print bm.astype(np.int) - bt.astype(np.int)
```

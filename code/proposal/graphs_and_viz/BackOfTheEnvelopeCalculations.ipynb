## Focal length requirement from sampling point of view

**Requirement:** 200 pixels across the iris diameter

If iris diameter = 10 mm, and we project 200 pixels on the iris, then

200 pixels in 10 mm

or 20 pixels/mm


20 pixels/mm can support a maximum of 10 cycles/mm of the object spatial frequency without aliasing. 

(just satisfy Nyquist criterion, i.e. 2 pixels/cycle with "proper" sampling phase)


=> minimum feature size $s$ (in cycles) on the iris (without aliasing) = 0.1 mm = 100 $\mu m$


**It is interesting that the above calculation matches up with the requirement specifications!! **

According to "Iris recognition–beyond one meter," by Matey and Kennell, 2009, 
The resolution (the authors call this as `iris pixel`) we require, as speciﬁed in the ISO standard for iris images [18] is of the order of 100 μ m = 0.01 cm [Ref# 18 is ISO/IEC JTC 1/SC 37 N 504, “Biometric Data Interchange Formats – Part 6: Iris Image Data”.]

### Other common methods of feature size calculation

####Matching spot-size to pixels

If we simply think about "matching" the pixels (projected on the iris) to the optical spot on the iris ... i.e. putting 200 resolvable spots on the iris, then the diameter of each spot is 10,000/200 = 50 micros .... Note that this is just a rough figure of resolution range (between 50 and 100 microns) required on the object side. 


So the basic formula for **maximum** feature size calculation is:

$$s_{max} = \frac{\text{iris_diameter}}{\text{pixels_on_iris} \times \text{cycles_per_pixel}}$$

where,

$$
\begin{align}
\text{cycles_per_pixel } & = 0.5 \text{ for no aliasing at Nyquist limit} \\\
\text{cycles_per_pixel } & = 1.0 \text{ for "matched" pixel-spot-size calculations} \\\
\text{cycles_per_pixel } & = 0.25 \text{ for no aliasing & 4 pixels per cycle} 
\end{align}
$$

### The relation between focal length (from the point of view of sampling), magnification, and object distance.

$$f_{p} = \frac{M_p u_o}{M_p + 1}$$

Let $p_x$ be the sensor pixel side (assumig square pixel for now) and let $s$ be the `iris pixel` or the minimum feature size.

The size of the projected pixel on the object is $p_x/M_p$, and we would like the $p_x/M_p \leq s$.

Then,

$$M_p = \frac{p_x}{s}$$

or

$$ f_p = \frac{p_x u_o}{p_x + s}$$

```python
from __future__ import division, print_function

def get_maximum_feature_size(iris_diam=10, pixels_on_iris=200, cycles_per_pixel=0.5):
    """return feature size"""
    return iris_diam/(cycles_per_pixel*pixels_on_iris)
    

def get_focalLength(pixel_length, min_iris_feature, obj_distance):
    """Returns focal length in mm and magnification
    
    Parameters
    ----------
    pixel_length  : side length of square pixel in length units
    min_iris_feature : minimum feature size of iris structure to resolve
                       likely without aliasing in length units
    obj_distance  :  standoff distance in length units
    
    Returns
    -------
    focal_length  : focal length in length units
    mag     :   lateral magnification
    
    Note 
    ----
    The focal length and mag calculations are length unit agonastic. Use
    same length units for all inputs.
    """
    mag = pixel_length/min_iris_feature
    focal_length = (mag/(mag + 1.0))*obj_distance
    return focal_length, mag
```

Note that in almost all iris acquisition subsystem design calculations, the lateral magnification (determined by the pixel size, number of pixels required on the iris and iris diameter) is constant. Based on the calculated, constant, lateral magnification the focal length is determined using the standoff distance. In such cases, the required focal length over distance varies proportional to the ratio of the standoff distances.

In other words,

$$\text{focal_length}_2 = \frac{\text{standoff_distance}_2}{\text{standoff_distance}_1} \times \text{focal_length}_1$$

## Optical spatial resolution vs. F-number

Let $x = \frac{\rho}{2\rho_o}$. Then $\rho \leq 2\rho_o \Rightarrow 0 \leq x \leq 1$

Then the diffraction limited OTF for the circular apreture may be written as:

$$
\mathcal{H}(\rho) = \frac{2}{\pi} \left[ arccos(x) - x\sqrt{1 - x^2}   \right]
$$


```python
# This plot is just to verify that the plot below is correct
fn = np.arange(1,17,0.1)
lamda = 850e-6  # mm 
r = 1.0/(lamda*fn)

fig = plt.figure(figsize(8,6))
ax = fig.add_subplot(111)
plot(fn, r)
ax.grid('on')

ax.set_xlabel('F-number')
ax.set_ylabel('Spatial cutoff-frequency (lppm)')
#ax.set_xscale('log')
#ax.set_yscale('log')

ax.margins(0)
```

### Calculation of resolution $f_0$ and $f_{60}$ at image or object planes for papers @ Lit Survey

```python
def getIPres(f_number, lamda=850e-6):
    """get resolution at image plane"""
    f0 = 1.0/(lamda*f_number)
    f60 = 0.31969151*f0
    return f0, f60

def getOPres_givenM(f_number, M, lamda=850e-6):
    """get resolution at object plane, given lateral magnification"""
    f0i, f60i = getIPres(f_number, lamda)
    return M*f0i, M*f60i
```

## Geometric Depth of Field calculation

The expression for the geometric depth of field (does not depend on the wavelength $\lambda$) is as follows:

$$DOF_{\text{geom}} = \frac{2 f^2 N c u^2 }{f^4 - N^2 c^2 u^2 }$$


Another approximation for the Geometric Depth of Field is as follows:

$$
DOF_{\text{geom}} = \frac{2 c N (u - f)^2}{f^2}
$$

The above equation is derived from the expression of geometric DOF given in 30.8 of Handbook of Optical Systems, vol 3, Gross.

```python
def get_geom_DOF(focal_length, Fnum, c, u, grossExpr=False):
    """Returns the geometric depth of field value
    
    Parameters
    ----------
    focal_length : focal length
    Fnum : F/# of the optical system
    c : circle of confusion in the same length-unit as `f`
    u : standoff distance in the same unit as `f`
    grossExpr : Use the gexpression from Gross' Handbook of Optical systems
    
    Returns
    -------
    dof : length of the DOF in the same unit as `f`
    """
    if grossExpr:
        dof = (2.0*c*Fnum*(u-focal_length)**2)/focal_length**2
    else:
        dof = (2.0*focal_length**2*Fnum*c*u**2)/(focal_length**4 - (Fnum*c*u)**2)
    return dof
```

## Axial resolution -- Depth of Focus from diffraction standpoint

#####Reference:

1.  Born and Wolf, Page 491
2.  Handbook of Optics, Physical Image formation, Vol -2 
3.  Handbook of Optical Systems. Aberration Theory, Vol-3, Gross, Page 126


**Definitions:**

"It is usual to regard a loss of about 20% in intensity at the center of the image patch as permissible" [1]. 

"The half diffractive optical focus depth is defined as the distance, within which the peak intensity has decreased to 80%" [3].

The expression for the depth of focus is as follows:

$$
\delta z = \pm \frac{6.4 \lambda}{\pi} N^2 \approx \pm 2 \lambda N^2 
$$

```python
import iutils.opticsutils.foptics as fou

def half_diffractive_focal_depth(Fnumber, lamda):
    """Returns half diffractive optical focal depth"""
    return 6.4*lamda*Fnumber**2 / np.pi


# Ensure that the following code is correct.... this code is partially correct...as it always uses the f-number, 
# instead of using the effective f-number. use the function from iutils.opticsutils.foptics 
def get_diff_DOF(focal_length, Fnum, standoff, lamda):
    """ """
    dz_half = half_diffractive_focal_depth(Fnum, lamda)
    image_plane = standoff*focal_length/(standoff - focal_length)
    dof_lim_f =  (image_plane - dz_half)*focal_length/((image_plane - dz_half) - focal_length)
    dof_lim_n =  (image_plane + dz_half)*focal_length/((image_plane + dz_half) - focal_length)
    dof = dof_lim_f - dof_lim_n
    return dof

# tests 1
Fnum = 10
lam = 500e-6 #mm
half_diff_focal_depth = half_diffractive_focal_depth(Fnum, lam)
print (half_diff_focal_depth)
```

The geometric DOF increases linearly with F-number. The diffraction based DOF increases exponentially with F-number. Also, it is really hard to compare the two plots (i.e. between the geometric DOF and the diffraction based DOFF) ... the geometric DOF is dependent on the pixel size of the sensor (and the coc) and it is not dependent on the wavelength of light. The diffraction based DOF is dependent on the wavelength and it doesn't depend on the pixel size. Obviously, for photographic images, where the "persieved sharpness" of an image is increased due to larger pixel size (and aliasing) the geometric DOF is appropriate. However, for measure of true DOF and true sharpness (presense of unaliased spatial frequencies), the geometric DOF is inappropriate.

## <font color='DarkBrown'>Simulation of iris image blurring due to defocus</font>

Note: For the first run in every session, execute the case of no defocus to store the PSF scaling factor.

```python
# Helper functions

def first_order_geoImageParameters(focalLength, objDistance, objHeight=None):
    """Returns first order approximations of image distance, image height, and lateral magnification
    using Gaussian lens equation.
    """
    imgHeight = None
    imgDistance = focalLength*objDistance/(objDistance - focalLength)
    latMag = imgDistance/objDistance
    if objHeight:
        imgHeight = latMag*objHeight
    return imgDistance, imgHeight, latMag

def incoherent_diffraction_limited_cutoff(wxp, zxp, lamda):
    """
    Returns the diffraction limited incoherent cutoff frequency in cycles/mm
    
    Parameters
    ----------
    wxp : float; exit pupil radius
    zxp : float; exit pupil distance
    lamda : float; wavelength
    
    Return
    ------
    diff_ltd_cutoff_freq
    
    """
    return (2.0*wxp)/(lamda*zxp)

def diff_ltd_sim_criterion_satisfied(wxp, zxp, lamda, L, num_samples):
    """verifies whether or not diffraction-limited imaging simulation is satisfied
    
    Parameters
    ----------
    wxp : float; exit pupil radius
    zxp : float; exit pupil distance
    lamda : float; wavelength
    L : float; image plane side (assigning physical units to the geometric image)
    num_samples : integer, number of samples along one dimension of the image, assuming square image 
    
    See: Computational Fourier Optics by David Voelz (equ. 7.17, & sec. 8.71 for details)
    """
    two_f0 = incoherent_diffraction_limited_cutoff(wxp, zxp, lamda)
    du = L/num_samples
    nyquist_freq = 1/(2.0*du)
    return two_f0 < nyquist_freq
```

To change the amount of defocus, change `obj_offset` below...use the multiplier in front of the number .... 0 for no defocus, -1 and 1.

```python
# Optics specificiation
lamda = 550e-6           # 550 nm
focal_length = 100       # mm
aperture_r =  10          # mm
obj_distance = 1300      # 1.3 meter
obj_height = 12          # 12 mm (diameter of the iris + some space)

# for perfect imaging ... no defocus
img_distance, img_height, lat_mag = first_order_geoImageParameters(focal_length, obj_distance, obj_height)
f_num_eff = img_distance/(2.0*aperture_r)

# *********
# Defocus
# *********
obj_offset =   0.0 # 18.0 # -1.0*25 #   minus means towards lens, plus means away from the lens
off_img_pos, off_img_ht, _= first_order_geoImageParameters(focal_length, obj_distance+obj_offset, obj_height)

w020 = fou.w020FromDefocus(aperture_r, img_distance, 
                           off_img_pos - img_distance, wave_length=lamda)

#zxp = off_img_pos              # distance of exit pupil from gaussian image point.
# I now think that the exit pupil distance should be the distance from the sensor ... because
# now the acutal wavefront will converge at a different position on the optical axis away from
# the sensor location. This will be the so called "aberrated" wavefront. 
zxp = img_distance              # distance of exit pupil from gaussian image point for no defocus. 
L = off_img_ht #img_height  # off_img_ht    # ideally this should be the off_img_ht ... using img_height instead ... no change in magnification

print "f-number: ", focal_length/(2.0*aperture_r)
print "effective f-number:", f_num_eff
print "image distance: ", img_distance, "mm"
print "image height: ", img_height, "mm"
print "defoucs-image distance:", off_img_pos, "mm"
print "defocus image height: ", off_img_ht, "mm"
print "delta_z: ", (off_img_pos - img_distance)*1000.0, "microns"
print "amount of defoucs w020 :", w020
```

```python
# Coherent transfer function and PSF
# ----------------------------------
# aberration coefficient
# waves of defocus defocus (w020 is in terms of wavelength) ... this needs to be multiplied by
# the wavelength
defocus_coeff = w020*lamda   

# create the frequency grid Fu, Fv, for the fourier transform
fu = tu.fft_freq_bin(M, order=1, Fs=1/du) # same as np.arange(-1/(2*du),1/(2*du), 1/L) used in the book.
fv = fu  # since it is a square grid with equal number of samples along each axis
Fu, Fv = np.meshgrid(fu, fv)
wxp = aperture_r
k = 2*np.pi/lamda

# wavefront (for the axial image point 0,0) 
W = fou.seidel_5(0, 0, -lamda*zxp*Fu/wxp, -lamda*zxp*Fv/wxp, wd=w020*lamda)

# coherent transfer function
H = siu.circR(np.hypot(Fu, Fv)/f0)*np.exp(-1j*k*W)

# the autocorrelation is the inverseFFT of the power-spectrum, which is the square
# of the fft of the function.
OTF = np.fft.ifft2(np.abs(np.fft.fft2(np.fft.fftshift(H))**2)) 
OTF = OTF/OTF.max() # since the maximum occurs at dc

# Incoherent imaging PSF 
A = 1.0  # TO DO: Have to fill-in the correct scaling factor 
psf = A*np.fft.fftshift(np.fft.fft2(H))
psf2 = np.abs(psf)**2.0
psf2 = psf2/np.sum(psf2)  # normalized the sum under the impulse response is 1

## Figures

# plots of the coherent transfer function and optical transfer function
if 0:
    # plot the coherent transfer function as a surface plot
    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(111,projection='3d')
    ax.plot_surface(Fu,Fv,H,rstride=5,cstride=5,linewidth=0.0,cmap=cm.RdYlBu_r,alpha=1.0,antialiased=True)
    ax.set_xlabel('Fu(cy/mm)')
    ax.set_ylabel('Fv(cy/mm)');
    ax.set_title('Coherent Transfer Function, H')
    
    # plot the OTF as a surface plot
    fig = plt.figure(figsize=(8,5))
    ax = fig.add_subplot(111,projection='3d')
    ax.plot_surface(Fu,Fv,np.abs(np.fft.fftshift(OTF)),
                    rstride=5,cstride=5,linewidth=0.0,cmap=cm.RdYlBu_r,alpha=1.0,antialiased=True)
    ax.set_xlabel('Fu(cy/mm)')
    ax.set_ylabel('Fv(cy/mm)');
    ax.set_title('Optical Transfer Function')
    
    plt.show()
    
# plot of the intensity PSF
fig = plt.figure(figsize=(5,5))
ax1 = fig.add_subplot(111)

# Determine how much of the array of the PSF to show
# may be determined using encircled energy ciriterion
T = 300
psf2_show = psf2[T:M-T, T:M-T]   # assming square
u_index_show = u[T:M-T]
v_index_show = v[T:M-T]
ax1.imshow(psf2_show**0.65, cmap=psfcm, 
           extent=[u_index_show.min(), u_index_show.max(), v_index_show.min(), v_index_show.max()])
ax1.set_title('Incoherent PSF')
plt.show()
```

```python
L_new = 500e-3
M_new = 1000
du_new = L_new/M_new
u_new = np.arange(-L_new/2, L_new/2, du_new)
v_new = u_new

beta = 0.8    # power used for contrast enhancement of the PSF


# create the frequency grid Fu, Fv, for the fourier transform
fu_new = tu.fft_freq_bin(M_new, order=1, Fs=1/du_new) 
fv_new = fu_new  
Fu_new, Fv_new = np.meshgrid(fu_new, fv_new)

# wavefront
W_new = fou.seidel_5(0, 0, -lamda*zxp*Fu_new/wxp, -lamda*zxp*Fv_new/wxp, wd=w020*lamda)

# coherent transfer function
H_new = siu.circR(np.hypot(Fu_new, Fv_new)/f0)*np.exp(-1j*k*W_new)

#plt.imshow(np.abs(H_new), 'gray')

# Incoherent imaging PSF (intensity PSF)
psf_new = np.fft.fftshift(np.fft.fft2(H_new))
psf2_new = np.abs(psf_new)**2.0

# store the scaling factor. This scaling factor is calculated only when there is no defocus
# then the same scaling factor is used for the other two cases. Only when there is no defocus
# the effect of multiplying with the scaling factor will result in PSF(0,0)==1
if obj_offset==0:
    psf_scaling = 1.0/psf2_new.max()

#psf2_new = psf2_new/np.sum(psf2_new)  # normalized the sum under the impulse response is 1
psf2_new = psf2_new*psf_scaling  # normalized PSF at dc is 1.... 


# plot of the intensity PSF
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

# Determine how much of the array of the PSF to show
# may be determined using encircled energy ciriterion
T = 431
psf2_show = psf2_new[T:M_new-T, T:M_new-T]   # assming square
u_index_show = u_new[T:M_new-T]
v_index_show = v_new[T:M_new-T]
im = ax.imshow(psf2_show**beta, cmap=psfcm, 
           extent=[u_index_show.min(), u_index_show.max(), v_index_show.min(), v_index_show.max()],
           vmax=1.0**beta, vmin=0)   

# Add a color bar on top
cbar = plt.colorbar(im, orientation='horizontal', fraction=.046, pad=0.04, drawedges=False)
cbar.outline.set_visible(False)
cbar.ax.tick_params(direction='out')

#ax.set_title('Incoherent PSF')
# remove ticks
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')

ext = u_index_show.max() - u_index_show.min()
ax.set_xlim(u_index_show.min(), u_index_show.max())
ax.set_ylim(v_index_show.max(), v_index_show.min())

# Put Scale on 
scale_ratio = 1/4.975  # proportion of scale w.r.t. axis width (try to keep the scale an integer)
scale_length = scale_ratio*(ext) # length of scale in units of data
scale_startx = ext/2 - scale_length - scale_length/5.0 # horizontal line
scale_endx = scale_startx+scale_length # horizontal line
scale_starty = ext/2 - (1/25.)*(ext) # horizontal line
scale_endy = scale_starty       # horizontal line
scale_bar_ht = (1/10.)*scale_length
ax.plot([scale_startx, scale_endx], [scale_starty, scale_endy], 'w', lw=1.8)
ax.plot([scale_startx, scale_startx], [scale_starty - scale_bar_ht/2.0, scale_starty + scale_bar_ht/2.0], 'w', lw=1.8)
ax.plot([scale_endx, scale_endx], [scale_starty - scale_bar_ht/2.0, scale_starty + scale_bar_ht/2.0], 'w', lw=1.8)
ax.text(scale_endx-1.05*scale_length, scale_starty-scale_bar_ht, 
             r'${} \, \mu m$'.format(np.round(scale_length*1000, decimals=2)), fontsize=15, color='w')

# Put lambda and F/# information
anno_x = -ext/2 + (1/50.)*(ext)
anno_y_line1 = scale_starty - scale_bar_ht
anno_y_line2 = scale_starty + scale_bar_ht - 0.0005
ax.text(anno_x, anno_y_line2, r'$\lambda = \,{lv}\,nm\,| \, F/\# = \, {fn:2.2f}$'.format(lv=lamda*1e6, fn=f_num_eff), 
         fontsize=16, color='#CCCCFF')
fig.tight_layout()

if bSaveFigure:
    saveFigure(fig, "problem_vis_03_psf_{}_objDist_2".format(obj_distance+obj_offset), DPI=100)
plt.show()
```

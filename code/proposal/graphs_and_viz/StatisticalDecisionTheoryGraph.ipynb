```python
# figure 7 and 11 of Daugman's paper
from __future__ import division
from scipy.stats import binom
from scipy import interpolate

# Figure (7) -- Statistical decision theory concept (using normal distributions) 
N_normal = 1000
n_normal = linspace(0,1,N_normal)
# Using the definition of Gaussian distribution
# as the norm function plots the standard normal
def my_gaussian(mu, sigma, x):
    dist = 1.0/(sigma*np.sqrt(2.0*np.pi))*np.exp(-(x-mu)**2 / (2.0*sigma**2))
    return dist

Y_normal1 = my_gaussian(0.3, 0.08, n_normal)
Y_normal2 = my_gaussian(0.5, 0.08, n_normal)


# Figure (11) -- Binomial distributions
N1 = 480   # Number of trials
p1 = 0.45  # 

N2 = 40
p2 = 0.084 

X1 = binom(N1, p1)
X2 = binom(N2, p2)

x1 = linspace(0,N1,N1+1)
x2 = linspace(0,N2,N2+1)

Y1 = X1.pmf(x1)
Y2 = X2.pmf(x2)

# Combine the two distributions
f = interpolate.interp1d(x2, Y2)
x2_new = linspace(0, N2, N1+1)
Y2_new = f(x2_new)

# Need to appropiately scale Y1 as the number of trials are different
Y1_scaled = 7.0*Y1   # the value 7.0 was selected by looking at the graph!!!! humm ...

# ===============
# Plotting
# ===============
# Create figure
if bSaveFigure:
    fig = plt.figure(facecolor='white', figsize=(14,5), dpi=100)
else:
    fig = plt.figure(facecolor='white', figsize=(14,5))
    
tickcol = '#111111'

#red_tintcol = (236/255.0, 88/255.0, 164/255.0)
#blue_tintcol = (154/255.0, 196/255.0, 131/255.0)

red_tintcol = (239/255.0,138/255.0,98/255.0)
blue_tintcol = (103/255.0,169/255.0,207/255.0)
red_tintcol2 = (178/255.0,24/255.0,43/255.0)
blue_tintcol2 = (33/255.0,102/255.0,172/255.0)

def common_ax_settings(axi):
    # Spines 
    axi.spines['right'].set_color('none')
    axi.spines['top'].set_color('none')
    axi.spines['left'].set_color('#888888')
    axi.spines['bottom'].set_color('#888888')
    axi.spines['bottom'].zorder = 22
    # Tick settings
    axi.xaxis.set_ticks_position('bottom') 
    axi.yaxis.set_ticks_position('left') 
    axi.xaxis.set_tick_params(direction='out', color=tickcol, pad=0.1)

# plot the Gaussian distributions
# ----------------------------
ax = fig.add_subplot(121)
ax.plot(n_normal, Y_normal1, color=red_tintcol, linewidth=1.85)
ax.plot(n_normal, Y_normal2, color=blue_tintcol, linewidth=1.85)

# Decision line/ Threshold line
C = 0.4
ax.vlines(C, 0.0, 5.6, color=(103/255.0, 0/255.0, 31/255.0), linewidth=1.5, linestyle='--', zorder=20)
ax.fill_between(x=n_normal, y1=Y_normal1, y2=Y_normal2, where=Y_normal1>Y_normal2,
                color=red_tintcol, hatch='', alpha=0.7)
ax.fill_between(x=n_normal, y1=Y_normal2, y2=Y_normal1, where=Y_normal2>Y_normal1,
                color=blue_tintcol, hatch='', alpha=0.7)

# fill the intermediate regions (error rate regions)
ax.fill_between(n_normal[n_normal<C], Y_normal2[n_normal<C], 
                color=blue_tintcol2, hatch='..', alpha=0.75)
ax.fill_between(n_normal[n_normal>=C], Y_normal2[n_normal>=C], where=Y_normal1[n_normal>=C]>Y_normal2[n_normal>=C],
                color=red_tintcol2, hatch='..', alpha=0.75)
ax.fill_between(n_normal[n_normal>=0.4], Y_normal1[n_normal>=0.4],
                color=red_tintcol2, hatch='..', alpha=0.75)

ax.set_xlim(0,1)
ax.set_ylim(0, 6) # Note that the distributions are not normalized
common_ax_settings(ax)

# x-labels (Hamming distance)
#xtic_locs = plt.MaxNLocator(11)
#ax.xaxis.set_major_locator(xtic_locs)
#ax.xticks = ax.get_xticks()
#ax.set_xticklabels(xticks, color=tickcol, fontsize=13, family='serif')
ax.xaxis.set_ticks([])

ax.set_yticks([])
ax.set_xlabel('Measure of similarity (D)', fontsize=13, family='sans-serif', color=(0.1,0.1,0.1), labelpad=22)
ax.set_ylabel('Frequency', fontsize=13, family='sans-serif', color=(0.1,0.1,0.1))
#ax.set_title('(a) Statistical Decision Theory', fontsize=15)
ax.text(0.33, 5.05, 'Criterion', fontsize=12, family='sans-serif',
        bbox=dict(facecolor='white',edgecolor='white',alpha=1.0), zorder=23)

ax.text(0.622, 1.5, 'Accept if D < Criterion \nReject if D > Criterion', fontsize=12,family='serif',
        linespacing=1.5, bbox=dict(facecolor='white',edgecolor=(0.85,0.85,0.85),alpha=0.5), zorder=23)

# Annotations
ax.annotate('Intra-class \nvariability', (0.26,4.5), (0.02,5.1), fontsize=12,family='serif',
            arrowprops=dict(arrowstyle='-',connectionstyle="arc3,rad=0.2",color=(0.1,0.1,0.1), linewidth=1.,
                            relpos=(1,0)), color=(0,0,0), zorder=20)
ax.annotate('Inter-class \nvariability', (0.54,4.5), (0.59,5.1), fontsize=12,
            arrowprops=dict(arrowstyle='-',connectionstyle="arc3,rad=-0.2",color=(0.1,0.1,0.1), linewidth=1.,
                            relpos=(0,0)), color=(0,0,0), zorder=20)


ax.annotate('True Accept', (0.27,2.3), (0.725,2.98), fontsize=12,family='serif',
            arrowprops=dict(arrowstyle='-',connectionstyle="arc,angleA=0,angleB=60,armA=40, armB=40,rad=0.0",
                            relpos=(0,0.5), color=(0.1,0.1,0.1), linewidth=0.9),
            color=(0,0,0), zorder=20)
ax.scatter([0.27],[2.3],c=['k'],zorder=10)  # arrow head
ax.annotate('True Reject', (0.55,2.2), (0.725,2.703), fontsize=12,family='serif',
            arrowprops=dict(arrowstyle='-',connectionstyle="arc,angleA=0,angleB=60,armA=40, armB=30,rad=0.0",
                            relpos=(0,0.5), color=(0.1,0.1,0.1), linewidth=0.9),
            color=(0,0,0), zorder=20)
ax.scatter([0.55],[2.2],c=['k'],zorder=10)  # arrow head
ax.annotate('False Reject', (0.42,1.0), (0.725,0.49), fontsize=12,family='serif',
            arrowprops=dict(arrowstyle='-',connectionstyle="arc,angleA=0,angleB=-60,armA=40, armB=24,rad=0.0",
                            relpos=(0,0.5), color=(0.1,0.1,0.1), linewidth=0.9),
            color=(0,0,0), zorder=20)
ax.scatter([0.42],[1.0],c=['k'],zorder=10)  # arrow head
ax.annotate('False Accept', (0.37,0.78), (0.725,0.19), fontsize=12,family='serif',
            arrowprops=dict(arrowstyle='-',connectionstyle="arc,angleA=0,angleB=-60,armA=40, armB=28,rad=0.0",
                            relpos=(0,0.5), color=(0.1,0.1,0.1), linewidth=0.9),
            color=(0,0,0), zorder=20)
ax.scatter([0.37],[0.78],c=['k'],zorder=10)  # arrow head

# plot the Binomial distributions
# ------------------------------
ax = fig.add_subplot(122)
ax.plot(Y1_scaled, color=blue_tintcol, linewidth=1.85)
ax.plot(Y2_new, color=red_tintcol, linewidth=1.85)
ax.fill_between(x1, Y1_scaled, color=blue_tintcol, hatch='', alpha=0.7)
ax.fill_between(x1, Y2_new, color=red_tintcol, hatch='', alpha=0.7)


ax.set_xlim(0, x1.max())
ax.set_ylim(0, 1.1*Y1_scaled.max())
common_ax_settings(ax)

# x-labels (Hamming distance)
hd_index = linspace(0,1, N1+1)
xtic_locs = linspace(0, N1, 11)
xtic_lbs = ['{:0.1f}'.format(hd_index[i]) for i in xtic_locs]
ax.set_xticklabels(xtic_lbs, color=tickcol, fontsize=11, family='sans-serif')
ax.set_xticks(xtic_locs)

ax.set_yticks([])
ax.set_xlabel('Hamming Distance (HD)', fontsize=13, family='sans-serif', color=(0.1,0.1,0.1))
ax.set_ylabel('Frequency', fontsize=13, family='sans-serif', color=(0.1,0.1,0.1))

# Annotations
ax.annotate('Distribution from \ncomparing same\nirides',(0.095*N1,0.20), (0.04*N1,0.24), fontsize=12, family='serif',
            arrowprops=dict(arrowstyle='-',connectionstyle="arc3,rad=-0.2", relpos=(0.25,0),
                            color=(0.1,0.1,0.1), linewidth=1.), color=(0,0,0), zorder=20)
ax.annotate('Distribution from \ncomparing different\nirides',(0.47*N1,0.20), (0.5*N1,0.24), fontsize=12,
            arrowprops=dict(arrowstyle='-',connectionstyle="arc3,rad=-0.2", relpos=(0.05,0),
                            color=(0.1,0.1,0.1), linewidth=1.),color=(0,0,0), zorder=20)
#ax.annotate('HD Criterion (0.32)',(0.32*N1,0.0), (0.25*N1,0.05), fontsize=12,
#            arrowprops=dict(arrowstyle='-|>',relpos=(0.05,0),
#                            color=(0.1,0.1,0.1), linewidth=1.25),color=(0,0,0), zorder=20)
ax.vlines(0.32*N1, 0.0, 0.22, color=(103/255.0, 0/255.0, 31/255.0), linewidth=1.5, linestyle='--', zorder=20)
ax.text(0.285*N1, 0.14, 'HD Criterion (0.32)', fontsize=13, family='sans-serif', zorder=20, rotation=90)

#
plt.figtext(0.086, 0.86, '(a)', fontsize=16, weight=545, color='k')
plt.figtext(0.51, 0.86, '(b)', fontsize=16, weight=545, color='k')

plt.show()

# Save figure, if turned on
saveFigure(fig, fig_name='statisticalDecisionTheory2', DPI=72)# =fig.get_dpi())
# comment: trying to save the figure at a different DPI setting is messing up the arrow line angles 
```

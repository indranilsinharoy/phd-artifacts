Summary
-------

**Advantage of the Log-gabor filter over the Gabor filter:**

1. The even symmetric filters of the Gabor functions have non-zero DC component whenever the bandwidth is greater than 1 octave [1, 2]
2.  While the maximum bandwidth of a Gabor filter is limited to approximately 1 octave, the Log-Gabor filters can have arbitrary bandwidth [1].
3.  If wide bandwidth along with maximal spatial localization is desired, the Gabor filters do not produce the most optimium solutions [1].

**Other motivations of using Log-Gabor filters:**

1.  Natural images are better coded by filters that have Gaussian transfer functions when viewed on the logarithmic frequency scale [3].

### The Log-Gabor filter

The transfer function of the Log-Gabor filter (in linear frequency scale) is [1,2]:


$$G(f) = exp\left(\frac{-[log(f / f_o)]^2}{2[log(\sigma/ f_o)]^2}\right) $$


where,

  $f$ : frequency ($0 \leq f < 1$, however it is evaluated for $0 < f \leq 0.5$)  

  $f_o$ : filter's center frequency 
  
  $\sigma$ : bandwidth of the filter.


#### A closer look at the Log-Gabor filter

$$G(f) = exp\left(\frac{-[log(f) -log(f_o)]^2}{2[log(\sigma/ f_o)]^2}\right) $$

#### Characteristics of the Log-Gabor filter

1.  To obtain "constant shape ratio filters" the term $\sigma/f_o$ must also be held constant for varying $f_o$ [1].
2.  The Log-Gabor filters, by definition, never have DC components [1].
3.  "The transfer function of the log Gabor function has an extended tail at the high frequency end. Field's studies of the statistics of natural images indicate that natural images have amplitude spectra that fall off at approximately $1/f$. To encode images having such spectral characteristics one should use filters having spectra that are similar." [Quoted from 1.]

```python
from __future__ import division, print_function
import numpy as np
```

```python
# Log-Gabor filter function
def logGabor(fo, sigma, N=50, sigmaOnf=False, fullSpec=True, freqBinOrder=0, fmat=None):
    """Returns the Log-Gabor filter tranfer function, Gw(f), calculated for positive
    frequencies (i.e. Gw = 0, for f <= 0)
    
    Parameters
    ----------
    fo    : center frequency
    sigma : bandwidth of the filter if `sigmaOnf` is False. If `sigmaOnf` if True, 
            then this parameter is interpreted as `simga/fo`, which too controlls
            the bandwidth
    N     : Number of samples to generate in the frequency vector (default = 50) 
            for the full spectrum, i.e. for 0 <= f < 1. Note that `f` is 
            normalized frequency. If `fmat` is passed, then `N` is ignored.
    sigmaOnf : The bandwidth may also be set by specifying `sigmaOnf` instead of 
            `sigma`. If this parameter is set to `True` then the parameter
            `sigma` is interpreted as `sigma/fo`.
    fullSpec : not relavent if `fmat` is passed. If true (default) and `fmat` is 
             not passed, then the length of `Gw` and `fmat` vectors is N, the 
             arrangement of frequency indices in `fmat` vector is determined by 
             `freqBinOrder`
             If `fullSpec` if `False`, then only the transfer function corresponding
             to [0,0.5) is returned.
    freqBinOrder : determines the arrangement of the frequency bins in the returned
             transfer function.
             0 (default) = DC, positive_frequency, negative_frequency
             1 = negative_frequency, DC, positive_frequency
             The Nyquist frequency is present in the first bin of the negative_frequency
             group, if N if even.
    fmat   : frequency axis (1-D or 2-D array) optional, if provided, then (a 1-D) 
            frequency index is not generated in the function. For 2-D log-Gabor 
            filter transfer filters, the function expects the `fmat` to be passed. 
    
    Returns
    -------
    Gw    : frequency response of the Log-Gabor filter
    fmat  : frequency vector (see above for details)
    
    """
    if fmat==None:      # since `fmat` is an ndarray, "if fmat:" syntax won't work
        if N%2: # odd
            fp = np.arange(0, 0.5, 1.0/N)
            fn = -1.0*fp[-1:0:-1]
        else:
            fp = np.arange(0, 0.5, 1.0/N)
            fn = np.arange(-0.5, 0, 1.0/N)
        f = fp               # Calculate only for the positive index.
    else:
        f = fmat.copy()      # since we are going to modify it inside, and don't
                             # want to change the original...
        f_full = f           # `f_full` is just a reference to `f` 
    zero_ind = np.argwhere(f==0.0)
    f[zero_ind] = 1.0       # to avoid log(0)
    
    if sigmaOnf:
        sof = sigma
    else:
        sof = sigma/fo
    
    # Calculate the transfer function for the positive frequencies
    numr = -1.0*(np.log(f/fo))**2
    deno = 2.0*(np.log(sof))**2
    Gw = np.exp(numr/deno)
    Gw[zero_ind] = 0.0      # set the value of the zero-freq point back to 0
    f[zero_ind] = 0.0
    Gw_Nyq = np.exp((-1.0*(np.log(0.5/fo))**2)/(2.0*(np.log(sof))**2))
    
    # Combine the spectrum
    if fmat==None:
        Gw_full = np.zeros(N)
        if freqBinOrder:
            f_full = np.hstack((fn, f))
            Gw_full[N//2:] = Gw
            if N%2==0:
                Gw_full[0] = Gw_Nyq
        else:
            f_full = np.hstack((f, fn))
            Gw_full[0:(N+1)//2] = Gw
            if N%2==0:
                Gw_full[(N+1)//2] = Gw_Nyq
    else:
        Gw_full = Gw
    
    if fullSpec:
        return Gw_full, f_full
    else:
        return Gw, f
```

#### Log-Gabor filters in the spatial domain

1.  One cannot directly construct the analytic expression for the log Gabor functions in the spatial domain because of the singularity of the log function at the origin [1].
2.  In order to ... "One is reduced to designing the filters in the frequency domain and then performing a numerical inverse Fourier Transform to see what they look like. Their appearance is similar to Gabor functions though their shape becomes much `sharper` as the bandwidth is increased. The shapes of log Gabor and Gabor functions are almost identical for bandwidths less than one octave. Shown below are three log Gabor filters of different bandwidths all tuned to the same centre frequency." [Quoted from 1]

```python
# Currently is this only for Even samples
def plotLogGaborImpulseFunction(sigma_over_fo=0.74, fo=0.08, N=200):
    # Create the Even and Odd symmetric filter centered spectrum
    Gw, f = logGabor(fo, sigma_over_fo, N, sigmaOnf=True, freqBinOrder=1)
    # N is even
    Gwr = Gw[N/2 + 1:] # Next-2-DC to Last-bin-before-Nyquist
    Gwr = Gwr[::-1]    # Nyquist to -ve Next-2-DC
    Gwf_evenSym = 0.5*Gw.copy()
    Gwf_evenSym[1:N/2] = 0.5*Gwr
    Gwf_oddSym = -0.5*Gw.copy()
    Gwf_oddSym[1:N/2] = 0.5*Gwr
    # Rectify the Nyquist frequency bin that was also multiplied by -0.5
    Gwf_oddSym[0] = -1.0*Gwf_oddSym[0]

    # Equivalent centered transfer function
    GWF_e = Gwf_evenSym - 0j*Gwf_oddSym
    lg_e = np.real(np.fft.ifft(np.fft.fftshift(GWF_e)))
    GWF_o = 0*Gwf_evenSym - 1j*Gwf_oddSym
    lg_o = np.real(np.fft.ifft(np.fft.fftshift(GWF_o)))
    
    ## PLOTS
    
    # Plot the odd and even symmetric filters
    fig1 = figure(figsize=(11.75,4))
    ax11 = fig1.add_subplot(121)
    ax12 = fig1.add_subplot(122)
    ax11.plot(f, Gwf_evenSym,color='b')
    ax12.plot(f, Gwf_oddSym,color='r')
    ax11.set_ylim(-0.52,0.52); ax12.set_ylim(-0.52,0.52)
    ax11.set_xlim(-0.5, 0.5); ax12.set_xlim(-0.5, 0.5);
    ax11.set_title('Even symmetric'); ax12.set_title('j*Odd Symmetric')
    ax11.set_xlabel('Frequency, f'); ax12.set_xlabel('Frequency, f'); 
    # Plot the
    fig2 = plt.figure(figsize=(10,6))
    ax21 = fig2.add_subplot(211)
    ax22 = fig2.add_subplot(212)
    ax21.plot(np.fft.fftshift(lg_e),'b') # not sure why this fftshift is required!!!
    ax21.plot(np.fft.fftshift(lg_o),'r')
    ax22.plot(np.fft.fftshift(lg_e + lg_o), 'g')
    ax21.set_title('Even (blue) and Odd (red) symmetric impulse response')
    ax21.set_xlabel('Samples, n')
    ax22.set_title('Combined impulse response')
    ax22.set_xlabel('Samples, n')
    fig2.tight_layout()
    plt.show()
```

"Given that we are now able to construct filters of arbitrary bandwidth and zero DC component the following question arises: What is the best bandwidth to use? One observation is that as bandwidth increases so too does the sharpness of the filter. Therefore, one constraint might be imposed by the maximum sharpness of the filter that we can effectively represent. ### 2-D example

#### Radial Log-Gabor filter

Effect of varying the center frequency while keeping $\sigma/f_o$ constant

```python
# Adapted from [1]
rows, cols = 500, 500
x, y = np.linspace(-cols/2, cols/2 -1, cols)/cols, np.linspace(-rows/2, rows/2 -1, rows)/rows
fx, fy = np.meshgrid(x,y) 
radius = sqrt(fx**2 + fy**2)
fo = [0.05, 0.1, 0.2]
sof = 0.74

txlo = [0, cols/4, cols/2, 3*cols/4, cols-1]
txlb = ['{:0.2f}'.format(x[i]) for i in txlo]
tylo = [0, rows/4, rows/2, 3*rows/4, rows-1]
tylb = ['{:0.2f}'.format(y[i]) for i in txlo]


fig = plt.figure(figsize=(16,4))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

for i, fc in enumerate(fo):
    Gw, _ = logGabor(fo[i], sof, fmat=radius, sigmaOnf=True)
    eval("ax{}.imshow(Gw,'gray', interpolation='none')".format(i+1))
    eval("ax{}.set_xticklabels(txlb, rotation='vertical')".format(i+1))
    eval("ax{}.set_xticks(txlo)".format(i+1))
    eval("ax{}.set_yticklabels(tylb, rotation='horizontal')".format(i+1))
    eval("ax{}.set_yticks(tylo)".format(i+1))
    eval("ax{}.set_title('fo = {}')".format(i+1, fo[i]))
    
plt.show()
```

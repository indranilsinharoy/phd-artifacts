A 2-D Gabor filter over spatial domain $(x,y)$ is represented as [1]:

$$G(x,y) =  e^{-2\pi i \left[u_0(x - x_0) + v_0(y -  y_0)\right]} e^{-\pi \left[ \frac{(x - x_0)^2}{\alpha^2} + \frac{(y - y_0)^2}{\beta^2}\right]}$$

where $(x_0, y_0)$ specify position in the image, $(\alpha, \beta)$ specify the effective width and length, and $(u_0, v_0)$ specify modulation, which has spatial frequency $\omega_0 = \sqrt{u_0^2 + v_0^2}$

The above Gabor filter can also be written as a quadrature pair of two odd symmetric and even symmetric 2D Gabor filters as:

$$G_{even}(x,y) = cos \left( 2 \pi \big[u_0(x - x_0) + v_0(y -  y_0)\big] \right) e^{-\pi \left[ \frac{(x - x_0)^2}{\alpha^2} + \frac{(y - y_0)^2}{\beta^2}\right]} $$

and

$$G_{odd}(x,y) = -i . sin \left( 2 \pi \big[u_0(x - x_0) + v_0(y -  y_0)\big] \right) e^{-\pi \left[ \frac{(x - x_0)^2}{\alpha^2} + \frac{(y - y_0)^2}{\beta^2}\right]} $$

```python
from __future__ import division
import numpy as np
from numpy import linspace, real, imag, cos, sin, pi, exp, abs
from numpy.fft import fft, fftshift, fftfreq
import matplotlib.pyplot as plt
```

```python
# A rough study
def gabor1d(x, xo, uo, alpha):
    gauss = (1.0/1)*exp(-pi*((x - xo)**2 / alpha**2))
    evensym = cos(2.0*pi*uo*x)
    ioddsym = - 1j*sin(2*pi*uo*x)
    csinu = evensym + ioddsym
    return gauss*csinu, gauss, csinu, evensym, ioddsym

N = 100
x = linspace(0, N-1, N) - N//2

xo = 0
uo = 0.1
alpha = 7

gt, gauss, csinu, evensym, ioddsym = gabor1d(x, xo, uo, alpha)

# Fourier transforms of the components 
EVENSYM = (1.0/N)*fft(evensym)
IODDSYM = (1.0/N)*fft(ioddsym)

GAUSS = (1.0/N)*fft(gauss)
CSINU = (1.0/N)*fft(csinu)

GTr = fft(real(gt))
GTi = fft(imag(gt))

GT = (1.0/N)*fft(gt)

# Frequency index
freq = fftfreq(N)

# Plots of the components in time domain
fig0 = plt.figure(figsize=(16,4))
ax01 = fig0.add_subplot(121)
ax02 = fig0.add_subplot(122)
ax01.plot(x, gauss)
ax02.plot(x, real(csinu), x, imag(csinu),'g')
#ax02.plot(x[75:125], real(csinu)[75:125], x[75:125], imag(csinu)[75:125],'g')
ax01.set_xlim(x[0], x[-1]); ax02.set_xlim(x[0], x[-1])
ax01.set_xlabel('Samples, n'); ax02.set_xlabel('Samples, n')
ax01.set_title('Gaussian function (gauss)', fontsize='x-large')
ax02.set_title('real(csinu)=evensym=cos(...) & imag(csinu)=ioddsym/j=-sin(...)', fontsize='x-large')

# Plot the Even and the Odd symmetric parts in frequency domain
# Note that botht the spectrum is REAL ... 
fig1 = plt.figure(figsize=(16,4))
ax11 = fig1.add_subplot(121)
ax12 = fig1.add_subplot(122)
ax11.stem(fftshift(freq), fftshift(EVENSYM), 'r')
ax12.stem(fftshift(freq), fftshift(IODDSYM), 'r')
ax11.set_xlim(-0.5, 0.5); ax11.set_ylim(-0.55, 0.55)
ax12.set_xlim(-0.5, 0.5); ax12.set_ylim(-0.55, 0.55)
ax11.set_xlabel('Normalized frequency, f') 
ax12.set_xlabel('Normalized frequency, f')
ax11.set_title('EVENSYM', fontsize='x-large')
ax12.set_title('IODDSYM', fontsize='x-large')

# plots of the components in frequency domain
fig2 = plt.figure(figsize=(16,4))
ax21 = fig2.add_subplot(121)
ax22 = fig2.add_subplot(122)
ax21.plot(fftshift(freq), fftshift(abs(GAUSS)))
ax22.stem(fftshift(freq), fftshift(CSINU),'r')
ax21.set_xlim(-0.5, 0.5)
ax22.set_xlim(-0.5, 0.5); ax22.set_ylim(0,1.2)
ax21.set_xlabel('Normalized frequency, f') 
ax22.set_xlabel('Normalized frequency, f')
ax21.set_title('abs(GAUSS)', fontsize='x-large')
ax22.set_title('CSINU = EVENSYM + IODDSYM', fontsize='x-large')


# Gabor function time and frequency domain
fig3 = plt.figure(figsize=(16,4))
ax31 = fig3.add_subplot(121)
ax32 = fig3.add_subplot(122)
ax31.plot(x, np.real(gt), 'b', x, np.imag(gt), 'r')
ax31.set_title("Real(gt(n)); Imaginary(gt(n))", fontsize='x-large')
ax31.set_xlim(x[0], x[-1])
ax31.set_xlabel('Samples')
ax32.plot(fftshift(freq), fftshift(abs(GTr)),'b')
ax32.plot(fftshift(freq), fftshift(abs(GTi)),'r--')
ax32.set_xlim(-0.5, 0.5)
ax32.set_title('|FFT(real(gt))|; |FFT(imag(gt))|', fontsize='x-large')


fig4 = plt.figure(figsize=(16,4))
ax41 = fig4.add_subplot(121)
ax42 = fig4.add_subplot(122)
ax41.plot(fftshift(freq), fftshift(abs(GT)))
ax41.set_title("abs(GT) = |FFT(gt)|", fontsize='x-large')
ax41.set_xlim(-0.5, 0.5)
ax41.set_xlabel('Normalized frequency, f')
ax42.plot(fftshift(freq), fftshift(GT),'b')
ax42.plot(fftshift(freq), fftshift(abs(GT)),'r--')
ax42.set_xlim(-0.5, 0.5)
ax42.set_title('GT (blue); abs(GT)', fontsize='x-large')
ax42.set_xlabel('Normalized frequency, f') 

plt.show()
```

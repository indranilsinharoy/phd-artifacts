## Normalization step

#### Normalization (mask generation code is still left)

```python
from __future__ import division, print_function
import cv2
from scipy import interpolate
from scipy.interpolate import griddata
from iutils.imageutils.imageUtils import cv2mpl

np.set_printoptions(precision=4, linewidth=120, suppress=True)


def simple_interp(xo, yo, image):
    """This is a simple nearest neighbour interpolation
    The result from this function match that of numpy's griddata
    if the method is 'nearest'
    """
    r,t = xo.shape
    newimage = np.zeros((r,t))
    for i in range(r):
        for j in range(t):
            try:
                newimage[i][j] = image[int(round(yo[i][j]))][int(round(xo[i][j]))]
            except IndexError:
                newimage[i][j] = nan
    
    return newimage
    

def interpolation_griddata(xo, yo, image):
    """
    Note
    See: http://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.griddata.html
    """
    # Create a flat array for the iris image
    rn, cl = image.shape
    rows, cols = np.meshgrid(range(cl), range(rn))
    flat_indices = array([list((i,j)) for i, j in zip(rows.flatten(), cols.flatten())])
    flat_image = image.flatten()
    grid_im = griddata(flat_indices, flat_image, (xo, yo), method='nearest')
    return grid_im


def normalize_iris(image, iris, pupil, rad_pixels, ang_div, pixels2exclude=1):
    """Function for unwarp the iris annulus from 
    
    Parameters
    ==========
    image          : the input eye image to extract iris data from   
    iris           : 3-tuple (iris_center_x, iris_center_y, iris_radius)
    pupil          : 3-tuple (pupil_center_x, pupil_center_y, pupil_radius)
    rad_pixels     : integer, radial resolution. It defines the vertical dimension of 
                     the unwarpped iris section.
    ang_div        : angular resolution (number of spokes to extract from the iris annulas. 
                     It defines the horizontal dimension of the unwarpped iris section
    pixels2exclude : odd integer, number of pixels to exclude at the pupillary boundary
                     and at the limbic boundary.
    
    Returns
    =======
    polar_array :

    """
    if not pixels2exclude % 2:
        pixels2exclude =  pixels2exclude + 1
    rad_pixels = rad_pixels + 2*pixels2exclude
    
    theta = linspace(0,2*np.pi, ang_div)

    # calculate the displacement of pupil center from the iris center
    ox = pupil[0] - iris[0]
    oy = pupil[1] - iris[1]
    
    # calculate alpha and beta
    sgn = -1 if ox <=0 else 1 
    if oy and not ox:
        sgn = 1
        
    a = ox**2 + oy**2
    phi = np.arctan(oy/ox) if ox else np.pi/2
    b = sgn*np.cos(np.pi - phi - theta)
    
    # calculate the radius around the iris as a function of angle
    r = np.sqrt(a)*b + np.sqrt(a*b**2 - (a - iris[2]**2))
    r = r - pupil[2]
    rmat = np.tile(r, (rad_pixels, 1))
    r1 = np.linspace(0, 1, rad_pixels)
    r1mat = np.tile(r1, (ang_div, 1))
    rmat = rmat*r1mat.T
    rmat = rmat + pupil[2]
    
    # Exclude the values at the boundary of the pupillary boundary
    # and the limbic boundary
    rmat = rmat[pixels2exclude:-pixels2exclude,:]
    
    # calculate the location of each data point around the circular iris
    # in the cartesian plane
    ctheta = np.cos(theta)
    stheta = np.sin(theta)
    xo = rmat*ctheta[np.newaxis, ...] + pupil[0]
    yo = -1*rmat*stheta[np.newaxis, ...] + pupil[1]

    # extract intensity values into the normalized polar representation
    # using interpolation
    polar_array = simple_interp(xo, yo, image)
    #polar_array = interpolation_griddata(xo, yo, image)
    # Generation of polar mask
    polar_mask = None
    return polar_array, polar_mask, xo, yo
```

## Encoding step

Each row of the unwrapped iris will be colvolved with 1-D Log-Gabor wavelets. This operation, done on each row of the unwrapped iris image, is performed in the frequecy domain.

```python
import numpy as np
from __future__ import division, print_function

# Log-Gabor filter function
def logGabor(fo, sigma, N=50, sigmaOnf=False, fullSpec=True, freqBinOrder=0, fmat=None):
    """Returns the Log-Gabor filter tranfer function, Gw(f), calculated for positive
    frequencies (i.e. Gw = 0, for f <= 0)
    
    Parameters
    ----------
    fo    : center frequency
    sigma : bandwidth of the filter if `sigmaOnf` is False. If `sigmaOnf` if True, 
            then this parameter is interpreted as `simga/fo`, which too controlls
            the bandwidth
    N     : Number of samples to generate in the frequency vector (default = 50) 
            for the full spectrum, i.e. for 0 <= f < 1. Note that `f` is 
            normalized frequency. If `fmat` is passed, then `N` is ignored.
    sigmaOnf : The bandwidth may also be set by specifying `sigmaOnf` instead of 
            `sigma`. If this parameter is set to `True` then the parameter
            `sigma` is interpreted as `sigma/fo`.
    fullSpec : not relavent if `fmat` is passed. If true (default) and `fmat` is 
             not passed, then the length of `Gw` and `fmat` vectors is N, the 
             arrangement of frequency indices in `fmat` vector is determined by 
             `freqBinOrder`
             If `fullSpec` if `False`, then only the transfer function corresponding
             to [0,0.5) is returned.
    freqBinOrder : determines the arrangement of the frequency bins in the returned
             transfer function.
             0 (default) = DC, positive_frequency, negative_frequency
             1 = negative_frequency, DC, positive_frequency
             The Nyquist frequency is present in the first bin of the negative_frequency
             group, if N if even.
    fmat   : frequency axis (1-D or 2-D array) optional, if provided, then (a 1-D) 
            frequency index is not generated in the function. For 2-D log-Gabor 
            filter transfer filters, the function expects the `fmat` to be passed. 
    
    Returns
    -------
    Gw    : frequency response of the Log-Gabor filter
    fmat  : frequency vector (see above for details)
    
    """
    if fmat==None:      # since `fmat` is an ndarray, "if fmat:" syntax won't work
        if N%2: # odd
            fp = np.arange(0, 0.5, 1.0/N)
            fn = -1.0*fp[-1:0:-1]
        else:
            fp = np.arange(0, 0.5, 1.0/N)
            fn = np.arange(-0.5, 0, 1.0/N)
        f = fp               # Calculate only for the positive index.
    else:
        f = fmat.copy()      # since we are going to modify it inside, and don't
                             # want to change the original...
        f_full = f           # `f_full` is just a reference to `f` 
    zero_ind = np.argwhere(f==0.0)
    f[zero_ind] = 1.0       # to avoid log(0)
    
    if sigmaOnf:
        sof = sigma
    else:
        sof = sigma/fo
    
    # Calculate the transfer function for the positive frequencies
    numr = -1.0*(np.log(f/fo))**2
    deno = 2.0*(np.log(sof))**2
    Gw = np.exp(numr/deno)
    Gw[zero_ind] = 0.0      # set the value of the zero-freq point back to 0
    f[zero_ind] = 0.0
    Gw_Nyq = np.exp((-1.0*(np.log(0.5/fo))**2)/(2.0*(np.log(sof))**2))
    
    # Combine the spectrum
    if fmat==None:
        Gw_full = np.zeros(N)
        if freqBinOrder:
            f_full = np.hstack((fn, f))
            Gw_full[N//2:] = Gw
            if N%2==0:
                Gw_full[0] = Gw_Nyq
        else:
            f_full = np.hstack((f, fn))
            Gw_full[0:(N+1)//2] = Gw
            if N%2==0:
                Gw_full[(N+1)//2] = Gw_Nyq
    else:
        Gw_full = Gw
    
    if fullSpec:
        return Gw_full, f_full
    else:
        return Gw, f
```

```python
def gaborconvolve(image, nscales=1, minWavelength=18,mult=1, sigmaOnf=0.5):
    """Function for convolving each row of an image with 1-D log-Gabor
    filters. This is actually implemented in the frequency domain
    
    Parameters
    ----------
    image  : the image to filter using gaborconvolve (ndarray)
    nscales       : number of filters to use in encoding (integer)
    minWavelength : base wavelength (float)
    mult          : multiplicative factor between each filter (float)
    sigmaOnf      : bandwidth parameter (float)
    
    Returns
    -------
    filteredImg_stack : ndarray of complex valued convolution results.
                        the `ndim` for this array is 2<=ndim<=3. If `nscales` 
                        is greater than 1, each of the filtered image
                        at different scales are in the 3rd dimension
    """
    rows, cols = image.shape
      
    #if cols%2: # odd
    #    cols = cols -1
    
    filteredImg_stack = np.empty((rows, cols, nscales), dtype=np.complex64)
    wavelength = minWavelength
    
    
    # Get the Log-Gabor filter transfer function
    for i in range(nscales):
        fo = 1.0/wavelength
        Gw = np.zeros(cols)
        fvec = Gw.copy()
        Gw, fvec = logGabor(fo, sigmaOnf, cols, sigmaOnf=True)
        for r in range(rows):
            imgRow = image[r,0:cols]
            imgFFT = np.fft.fft(imgRow)
            filteredImg_stack[r,:,i] = np.fft.ifft(imgFFT*Gw)
        wavelength = wavelength*mult
    return filteredImg_stack
```

```python
def encodeIris(polar_array, noise_array, nscales=1, minWavelength=18,mult=1, sigmaOnf=0.5):
    """Returns an iris code template and mask from the normalized 
    iris image.
    
    Parameters
    ----------
    polar_array   : normalized iris region (ndarray)
    noise_array   : corresponding normalized noise region map (ndarray)
    nscales       : number of filters to use in encoding (integer)
    minWavelength : base wavelength (float)
    mult          : multiplicative factor between each filter (float)
    sigmaOnf      : bandwidth parameter (float)
    
    Returns
    -------
    template      : the binary iris biometric template (ndarray)
    mask          : the binary iris noise mask (ndarray)
    
    Note
    -----
    This function was slightly modified from Masek's implementation. In this
    function, the complete template is formed by stacking the sub-templates
    of each scale one after the other. 
    """
    filteredImg_stack = gaborconvolve(polar_array, nscales, minWavelength, mult, sigmaOnf)
    
    rows, cols = polar_array.shape
    template = np.zeros((rows, cols*2*nscales))
    mask = template.copy()
    
    for i in range(nscales):
        # Phase quantizations for every scale
        filt_data = filteredImg_stack[:,:,i]
        H1 = np.real(filt_data) > 0
        H2 = np.imag(filt_data) > 0
        # create noise mask for locations where the amplitude
        # is close to zero
        H3 = np.abs(filt_data) < 0.0001
        
        # create per scale template 
        template_curr_scale = np.zeros((rows,2*cols))
        template_curr_scale[:,::2] = H1
        template_curr_scale[:,1::2] = H2
        # place into main template container array
        template[:,cols*2*i:cols*2*(i+1)] = template_curr_scale
        
        # noise_mask
        # TO DO: OR the noise_array
        mask_curr_scale = np.zeros((rows,2*cols))
        mask_curr_scale[:,::2] = H3
        mask_curr_scale[:,1::2] = H3
        # place into main mask container array
        mask[:,cols*2*i:cols*2*(i+1)] = mask_curr_scale
    return template, mask
```

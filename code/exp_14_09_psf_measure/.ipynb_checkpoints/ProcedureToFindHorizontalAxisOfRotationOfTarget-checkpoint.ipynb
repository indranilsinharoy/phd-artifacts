Here, I would like to investiage some methods that will automatically calculate the tilt axis from the homography. In the following section, H1 is the homography calculated from the uncropped images, and H2 from cropped images. 

### Possible method 1: Try to estimate the euclidean distance between points and its transform for every point

```python
# Homography from the whole, uncropped image
H1 = [[ 9.43613054e-01, -6.51706101e-02,  1.84603759e+02],
      [ 2.58720236e-03,  6.57586485e-01,  1.00856861e+03],
      [-3.72715493e-08, -1.91863241e-05,  1.00000000e+00]]
```

```python
# cropping parameters
rs, re, cs, ce = 2066, 3915, 2237, 4987,
```

```python
# Homography from cropped image
H2 = np.array([[ 9.80340197e-01, -2.33546358e-02, 2.25058985e+01],
              [ 2.78680329e-03,  7.29828884e-01, 9.63275245e+01],
              [-3.64917282e-08, -2.00271365e-05, 1.00000000e+00]])
```

```python
def dist(x, y, H):
    """function to find the 2-norm distance between a point and 
    its transform ie. sqrt[(x-x')^2 + (y-y')^2]
    """
    fp = np.array([[x], [y], [1]])
    tp = np.dot(H, fp)
    tp = tp/tp[-1]
    return slinalg.norm(fp-tp)
```

Using the homography estimated using the whole image

```python
fpx = np.linspace(cs, ce, ce-cs, endpoint=False)
fpy = np.linspace(cs, re, re-rs, endpoint=False)      
x, y = np.meshgrid(fpx, fpy)
d = dist(x, y, H1)
```

```python
fig, ax = plt.subplots(1,1)
pl = ax.imshow(d)

N, M = ce - cs, re - rs
xTickLocs = [0, N/4, N/2, 3*N/4, N-1]
xTickLabl = ['{}'.format(int(fpx[i])) for i in xTickLocs]
ax.set_xticklabels(xTickLabl)
ax.set_xticks(xTickLocs)

yTickLocs = [0, M/4, M/2, 3*M/4, M-1]
yTickLabl = ['{}'.format(int(fpy[i])) for i in yTickLocs]
ax.set_yticklabels(yTickLabl)
ax.set_yticks(yTickLocs)

plt.colorbar(pl, fraction=0.035)
plt.show()
```

The figure, seems to suggest that the row number is around 3890. However, that is not correct, as it should have been around 2930.

Using the homography calculated using the cropped image

```python
fpx = np.linspace(0, ce-cs, ce-cs, endpoint=False)
fpy = np.linspace(0, re-rs, re-rs, endpoint=False)
x, y = np.meshgrid(fpx, fpy)
d = dist(x, y, H2)
```

### Second method -- using the eigen vectors of homography matrix

The idea is that the axis about which the rotation(s) happen shall not be changed by the homography. Then, the such a line is really an eigen vector of the homography matrix. 

```python
import scipy.linalg as _linalg 
```

```python
#The normalized right eigenvector corresponding to the eigenvalue ``w[i]`` is the column ``vr[:,i]`
w, vr = _linalg.eig(H2)
```

```python
# eigen values
w
```

```python
# eigen vectors
vr
```

Visual inspection of the lines (eigen vectors)

```python
def get_line(a, b, c, xlim):
    """
    a, b, c defines the line ax + by + c = 0
    """
    x = np.linspace(xlim[0], xlim[1])
    y = -(a/b)*x - (c/b)
    return x, y

x1, y1 = get_line(*vr[:,0].tolist(), xlim=(0, ce-cs))
x2, y2 = get_line(*vr[:,1].tolist(), xlim=(0, ce-cs))
x3, y3 = get_line(*vr[:,2].tolist(), xlim=(0, ce-cs))

fig, ax = plt.subplots(1, 1, figsize=(10,8))
#mpu.set_spines(axes=ax, stype='center_axes')
ax.plot(x1, y1, x2, y2, x3, y3)
ax.legend(['vr0, w={}'.format(w[0]), 
           'vr1, w={}'.format(w[1]), 
           'vr2, w={}'.format(w[2])])
plt.show()
```

The axis, is probably the shortest length line between the image boundaries ... here we consider only the vertical boundaries (i.e. between xlim and ylim)

```python
def get_line_length(a, b, c, col1, col2):
    """
    a, b, c defines the line ax + by + c = 0
    col1 is the first column (x-coordinate)
    col2 is the second column (x-coordinate)
    """
    # (col1, y1) will be the coordinates of the line
    # at the first boundary, and (col2, y2) will be the
    # coordinates of the line at the second boundary
    y1 = -(a/b)*col1 - (c/b)
    y2 = -(a/b)*col2 - (c/b)
    dist = np.sqrt((col2 - col1)**2 + (y2 - y1)**2)
    return dist, y1, y2

dist1, rowStart1, rowEnd1 = get_line_length(*vr[:,0].tolist(), col1=0, col2=ce-cs)
dist2, rowStart2, rowEnd2 = get_line_length(*vr[:,1].tolist(), col1=0, col2=ce-cs)
dist3, rowStart3, rowEnd3 = get_line_length(*vr[:,2].tolist(), col1=0, col2=ce-cs)
```

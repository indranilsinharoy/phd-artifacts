Analysis of PSFs 

```python
# Expected spot size for the object distance of 3000 mm, f=180 mm
flength = 180
objDist = 3000.0
wl = 550e-6
fnum = 16

m = eiu.mag(f=flength, u=objDist)
fnumEff = fou.effective_fnumber(fnum, m=m)

# Diffraction spot size @ Image space
sInf = 2.44*wl*fnum
print("Diffraction spot size @ infinite conjugate", sInf*1000, "microns")
sFinite = 2.44*wl*fnumEff
print("Diffraction spot size @ finite conjugate", sFinite*1000, "microns")

# How many pixels in the spot-size
pp = 6.0e-3  # Sinar 86H eVolution 
#pp = 2.4e-3  # Imaging Source, 
print("Fractional number of pixels assuming diffration limited performance:", sFinite/pp)

# Spot size in object space
sFiniteObjSpace = sFinite/m
print("Spot size in object space: ", sFiniteObjSpace*1000, "microns")

# Detector pixel in object space
print("Detector pixel in object space", pp*1000/m, "microns")

# Source pixel size
sourcePixelSizex = 40e-6 # microns
sourcePixelSizey = 90e-6 # microns
```

```python
imf_0 = "C:\\thesis_images\\2014_09_30_psfmeasurement\\zero_degrees\\CentralDot.tiff"
bff_0 = "C:\\thesis_images\\2014_09_30_psfmeasurement\\zero_degrees\\DarkFrame.tiff"
# Set 2
imf_1 = "C:\\thesis_images\\2014_09_30_psfmeasurement\\40_degrees\\CentralDot.tiff"
bff_1 = "C:\\thesis_images\\2014_09_30_psfmeasurement\\40_degrees\\DarkFrame.tiff"
```

```python
img0 = ndimage.imread(imf_0).astype(np.int32) - ndimage.imread(bff_0).astype(np.int32)
img1 = ndimage.imread(imf_1).astype(np.int32) - ndimage.imread(bff_1).astype(np.int32)
```

```python
# Normalize values between 0 and 1
img0 = apu.rescale(img0, low=0, high=1)
img1 = apu.rescale(img1, low=0, high=1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,10))
ax1.imshow(img0[170:184, 1277: 1290], interpolation='none')
ax2.imshow(img1[560:574, 1248: 1261], interpolation='none')
plt.show()
```
